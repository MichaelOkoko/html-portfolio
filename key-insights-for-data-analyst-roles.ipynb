 # Cleaning the Data

```python
# Drop unused columns
df_analyst.drop(["index", "extensions","job_id","thumbnail","search_term", "posted_at", "date_time","commute_time", "search_location"], axis=1, inplace=True)
df_analyst.columns
```

```python
# Check for duplicated rows
duplicated_rows = df_analyst.duplicated(subset=["title", "company_name", "description"], keep =False)
print(f"There are {duplicated_rows.count()} duplicated job postings.")

#if the job title, company name, and job description are identical the row is a duplicate 
cleaned_df = df_analyst.drop_duplicates(subset=["title", "company_name", "description"])
# Only first occurence of each duplicate is removed 
```

# Data Analysis and Visualization

**Lets Observe the elements in our dataset!**

```python
# Take the Job title name and tally its occurences,and sort higest to lowest
Job_title_total = cleaned_df.title.value_counts().sort_values(ascending = False)
# Store the top 5 most prevelant jobs
Top_Jobs = Job_title_total.nlargest(5)
Other_Jobs = Job_title_total.iloc[5:].sum()
# Combine top 5 jobs and other category into a pandas series
All_Jobs = pd.concat([Top_Jobs, pd.Series({'other': Other_Jobs})])

# Divide each of the top jobs by the total amount of jobs to get percentage, and set index of df to be same as All_Jobs
Job_title_total = pd.DataFrame({'Amount': All_Jobs.values, "Percentage": All_Jobs / sum(All_Jobs)*100 }, index = All_Jobs.index)
Job_title_total['Percentage'] = Job_title_total["Percentage"].map('{: .2f}%'.format)
#format percatages as floats with two decimals 

# Create chart
display(Job_title_total.style.background_gradient(cmap = "plasma"))
```

💡💡Here we can observe the most prevelant roles in our dataset, the frequency and its proportion in relation to the entire dataset. 

💡💡*** It is important to note*** In the 80% of 'other', there are the same roles of the top 5 with a character sequence that does not match exactly (ie.  Buisness Data Analyst, Marketing Data Analyst etc..)

🗣️🗣️ Not all Data Analyst role are the same. Each position represents a distinct stage along the career path from junior to senior level, each with its unique set of responsibilities and skill levels. Lets unravel the intracacies that differentiate these roles.


```python
# use boolean mask to filter and extract tuples of Jr, Mid, and Sr positions
junior = ["Jr", "Junior"]
Jr_mask = cleaned_df['title'].str.contains('|'.join(junior), case=False)
Mid_mask = cleaned_df['title'].str.contains('mid', case=False)
Senior = ['Senior', 'Sr']
Sr_mask = cleaned_df['title'].str.contains('|'.join(Senior), case=False)
# If string is contained it will create a True or False boolean in that row

#collect filtered rows by senority lvl, and store into new df
filtered_Jr = cleaned_df[Jr_mask]
filtered_Mid = cleaned_df[Mid_mask]
filtered_Senior = cleaned_df[Sr_mask]
```

```python
filtered_Jr.head()
```

**We observe only Junior Data Analyst roles!**

```python
#Find the number of Jr, Mid, Sr data analyst roles
Jr_roles = len(filtered_Jr)
Mid_roles = len(filtered_Mid)
Sr_roles = len(filtered_Senior)

data_roles = {'Senoirty of Role': ['Jr Level', 'Mid-level', 'Sr Level'],
      '# of Jobs': [Jr_roles, Mid_roles, Sr_roles]}
All_roles = pd.DataFrame(data_roles)

display(All_roles.style.background_gradient(cmap = "plasma"))
```

💡💡 We can see the amount of roles in our dataset that indicate either Jr, Mid, or Sr level senority. The rest of the roles do not specify.

🗣️🗣️ Knowing this information we can draw key insights into similarities and descrepincies within the varying levels. Below we will answer, how the desired level of education changes with senority, and more.

```python
# Roles requiring a Bachelors, Masters, or Docterate
# Search job decription to for the degree type
bachelor = (cleaned_df['description'].str.contains('bachelor', case=False)) 
masters = (cleaned_df['description'].str.contains('master', case=False)) 
docterate =(cleaned_df['description'].str.contains('phd', case=False)) 

#subset datasets filtered only those with degree types and roles of same senority
Jr_bach = filtered_Jr[bachelor]
Jr_mas = filtered_Jr[masters]
Jr_pos = filtered_Jr[docterate]

Mid_bach = filtered_Mid[bachelor]
Mid_mas = filtered_Mid[masters]
Mid_pos = filtered_Mid[docterate]

Sr_bach = filtered_Senior[bachelor]
Sr_mas = filtered_Senior[masters]
sr_pos = filtered_Senior[docterate]

# Obtain the percentage of roles Jr/Mid/Sr requiring differrent degree types  
b_jr_count = (len(Jr_bach) / Jr_roles)*100
m_jr_count = (len(Jr_mas) / Jr_roles) *100
p_jr_count = (len(Jr_pos) / Jr_roles) *100

b_mid_count = (len(Mid_bach) / Mid_roles) *100
m_mid_count = (len(Mid_mas) / Mid_roles) *100
p_mid_count = (len(Mid_pos) /Mid_roles) *100

b_sr_count = (len(Sr_bach) /Sr_roles)*100
m_sr_count = (len(Sr_mas) /Sr_roles)*100
p_sr_count = (len(sr_pos) /Sr_roles)*100


# Store counts of roles in dictionaries
junior = {'Bachelors': b_jr_count, 'Masters': m_jr_count, 'PhD': p_jr_count}
mid = {'Bachelors': b_mid_count, 'Masters': m_mid_count, 'PhD': p_mid_count}
senior = {'Bachelors': b_sr_count, 'Masters': m_sr_count, 'PhD': p_sr_count}

# Create 3 list to store the counts of roles for each level
degrees = ['Bachelors', 'Masters', 'PhD']
junior_counts = [junior[degree] for degree in degrees]
mid_counts = [mid[degree] for degree in degrees]
senior_counts = [senior[degree] for degree in degrees]

# Set the positions of the bars on the x-axis (set up triple bar graph)
x = np.arange(len(degrees))

# Set the width of the bars
width = 0.25

# Create the bar plot
plt.bar(x - width, junior_counts, width, label='Junior')
plt.bar(x, mid_counts, width, label='Mid')
plt.bar(x + width, senior_counts, width, label='Senior')

# Add labels and title
plt.xlabel('Degree')
plt.ylabel('Percentage')
plt.title('Education Requirements for Varying Seniority Levels of Data Analyst Roles')
plt.xticks(x, degrees)
plt.legend()

# Show plot
plt.show()
```

💡💡 It is shown that holding a Bachelor's degree is typically the benchmark for data analyst jobs and as the senority of the role increases its is more strictly required. In fact higher level degrees may be saught after among higher level roles.

**It is important to note:** In this dataset there were more Jr and Sr roles indicated, than mid level positions. Therefore a few mid-level roles may skew the trend that we expect to see.


```python
#ast to evaluate python expressions as literal syntax trees/ data structures
import ast
# create free list of all skills from description_tokens
cleaned_df['skills'] = cleaned_df['description_tokens'].apply(lambda x: list(set(ast.literal_eval(x))))
# plot the skills needed
# Converts each list element into seperate row, flattens column and counts occurence of unique skills
skill_counts = pd.DataFrame(cleaned_df['skills'].explode()).value_counts().reset_index()
# skill_count contains skills and frequency values

#splice top 10 skills
top_skills = skill_counts[:10]
 
#plot graph
plt.bar(x = top_skills['skills'],
      height = top_skills['count'], color = 'mediumpurple')
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Top 10 Desired Skills in Data Roles", fontsize = 16, fontweight = "bold")
plt.ylabel("Frequency", fontsize = 13)
plt.xlabel("Skills")
plt.show() 
```

💡💡 From this bar graph we can see the top tools, softwares, and languages most desired in data analyst roles. With no suprise leading the pack is SQL (structured Query Language). SQL is  most commonly used for manging and manipulating relational databases, making it essential for data analyst. It is then followeed by other important skills such in Excel, Pyhton, tablueu, Power BI and R.

🗣️🗣️ You will hear Python and R discusses greatly in data analytics/scinece, but which is better? Lets discuss.

```python
# Create datasets for pytnon (in jr, mid, sr )
python_jr = filtered_Jr['description_tokens'].str.contains('python', case=False).sum()
python_mid = filtered_Mid['description_tokens'].str.contains('python', case=False).sum()
python_sr = filtered_Senior['description_tokens'].str.contains('python', case=False).sum()

# Pattern is set to find the individual character 'r' and not within a string
pattern = r'\br\b'
# Create datasets for R (in jr,mid, sr)
r_jr = filtered_Jr['description_tokens'].str.contains(pattern, case=False).sum()
r_mid = filtered_Mid['description_tokens'].str.contains(pattern, case=False).sum() 
r_sr = filtered_Senior['description_tokens'].str.contains(pattern, case=False).sum() #& cleaned_df['description'].str.contains('degree', case=False))


# Store the percentage of jobs requiring  python or R among Jr,Mid,Sr
py_jr_count = (python_jr/ Jr_roles) *100
py_mid_count  = (python_mid/ Mid_roles) *100
py_sr_count = (python_sr/ Sr_roles) *100

r_jr_count = (r_jr/ Jr_roles) *100
r_mid_count = (r_mid/ Mid_roles) *100
r_sr_count = (r_sr/ Sr_roles) *100

# Store counts of roles in dictionaries
jr_ = {'Python': py_jr_count, 'R': r_jr_count}
mid_ = {'Python': py_mid_count, 'R': r_mid_count}
sr_ = {'Python': py_sr_count, 'R': r_sr_count}

# Create 3 list to store the counts of roles for Python and R 
tools = ['Python', 'R']
junior_tools = [jr_[tool] for tool in tools]
mid_tools = [mid_[tool] for tool in tools]
senior_tools = [sr_[tool] for tool in tools]

# Set the positions of the bars on the x-axis
n = np.arange(len(tools))

# Set the width of the bars
width = 0.25

# Create the bar plot
plt.bar(n - width, junior_tools, width, label='Junior')
plt.bar(n, mid_tools, width, label='Mid')
plt.bar(n + width, senior_tools, width, label='Senior')

# Add labels and title
plt.xlabel('Progamming Language')
plt.ylabel('Percentage')
plt.title('Python vs R: Most Used Programming Language in Data Roles')
plt.xticks(n, tools)
plt.legend()

# Show plot
plt.show()
```

💡💡 This visualization clearly shows that at all levels of senority pyhton is the more desired programming Language. While both can get the job done python has versatility through its extensive libraries (Numpy, Pandas, Matplotlib etc.) and capabilities in data vizualiztions. Pyhton has a larger commmunity which also influences why it has been adopted as the industry standard. 

```python
# Create datasets for Power BI (in jr, mid, sr )
power_jr = filtered_Jr['description_tokens'].str.contains('power_bi', case=False).sum()
power_mid = filtered_Mid['description_tokens'].str.contains('power_bi', case=False).sum()
power_sr = filtered_Senior['description_tokens'].str.contains('power_bi', case=False).sum()

# Create datasets for Tableau (in jr, mid, sr )
tableau_jr = filtered_Jr['description_tokens'].str.contains('tableau', case=False).sum()
tableau_mid = filtered_Mid['description_tokens'].str.contains('tableau', case=False).sum() 
tableau_sr = filtered_Senior['description_tokens'].str.contains('tableau', case=False).sum() #


# Store the percentage of jobs requiring  Power BI or Tableau among Jr,Mid,Sr
pow_jr_count = (power_jr/ Jr_roles) *100
pow_mid_count  = (power_mid/ Mid_roles) *100
pow_sr_count = (power_sr/ Sr_roles) *100

tab_jr_count = (tableau_jr/ Jr_roles) *100
tab_mid_count = (tableau_mid/ Mid_roles) *100
tab_sr_count = (tableau_sr/ Sr_roles) *100

# Store counts of roles in dictionaries
jr_pt = {'PowerBI': py_jr_count, 'Tableau': r_jr_count}
mid_pt = {'PowerBI': py_mid_count, 'Tableau': r_mid_count}
sr_pt = {'PowerBI': py_sr_count, 'Tableau': r_sr_count}

# Create 3 list to store the counts of roles for Python and R 
softwares = ['PowerBI', 'Tableau']
junior_soft = [jr_pt[software] for software in softwares]
mid_soft = [mid_pt[software] for software in softwares]
senior_soft = [sr_pt[software] for software in softwares]

# Set the positions of the bars on the x-axis
a = np.arange(len(softwares))

# Set the width of the bars
width = 0.25

# Create the bar plot
plt.bar(a - width, junior_tools, width, label='Junior')
plt.bar(a, mid_tools, width, label='Mid')
plt.bar(a + width, senior_tools, width, label='Senior')

# Add labels and title
plt.xlabel('Software')
plt.ylabel('Percentage')
plt.title('PowerBI vs Tableau: Preferred Data Vizualizaiton Software in Data Roles')
plt.xticks(n, softwares)
plt.legend()

# Show plot
plt.show()
```

💡💡 Across Junior, Mid and Senior roles, Power BI is found to be more prevelant. Both Power BI and Tableau are data visualiztions softwares essential in the industry for making dynamic dashboards and visuals. Both are commonly used and when proficient in one it, the skills can be transferable.

```python
# Fill empty cells with Not specified
cleaned_df['work_from_home'] = cleaned_df['work_from_home'].fillna('Not Specified')

# Create figure dimensions and set colours
plt.figure(figsize =(6,6))
colours = ['c', 'm']

# Create a Donut chart
plt.pie(cleaned_df['work_from_home'].value_counts(), labels = cleaned_df['work_from_home'].value_counts().index,
        autopct='%1.1f%%', startangle= 90, colors = colours, wedgeprops=dict(width=0.35))
plt.title("Work From Home Distribution")

plt.show()
```

🗣️🗣️ Let's now see the position types found in our dataset.

```python
job_terms = ["Full-time", "Part-time", "Contractor","Internship"]

# Extracting and counting the occurrences of each job term in schedule type
job_term_count = {attr: cleaned_df['schedule_type'].apply(lambda x: attr in str(x)).sum() for attr in job_terms}
# Lambda function checks if job term is present in the as string in each cell of schedule_type column
job_term_count

# Iterates job term, for each job type looks up count and add to job_types_values list
job_types_values = [job_term_count[job_type] for job_type in job_terms]
# ^^ contains count values corresponding to each job attribute term in 'job_terms'

# Set size of figue, then plot pie chart
plt.figure(figsize=(10, 7))
plt.pie(job_types_values, labels=job_terms, autopct='%1.1f%%', startangle=110, colors=['mediumpurple', 'red', 'deepskyblue', 'gold', 'lightsalmon'])
plt.title('Distribution of Job Types')
plt.show()
```

💡💡 Here is a pie chart showcasing the distribution of the various job types in the dataset. It is clearly seen that the majority of postings are for full time positions (69.5%). The next most common are contract employment followed by part-time then internship.

🗣️🗣️ Lets talk Money!!

```python
#prevelance of jobs w salaries
#group the dataframe by location column and calculates mean of max and min salaries and takes top 20
df=cleaned_df.groupby('location')[['salary_max','salary_min']].mean().sort_values(['salary_max','salary_min'],ascending=False).head(20)

# Use Plotly imported as go (graph Objects) visualizse avg min and max 
fig = go.Figure()

fig.add_trace(go.Bar(x=df.index,y=df['salary_min'],name='Minimum salary'))
fig.add_trace(go.Bar(x=df.index,y=df['salary_max'],name='Maximum Salary'))

fig.update_layout(title='Top 20 cities with their Minimum and Maximum Salaries',barmode='stack')
# Plot top 20
fig.show()
```

💡💡 Above shows the wide range in distribution of salaries across various cities in the United States. According to our dataset the average minimum salary in Rolla Missoura is 109.6k and the average maximum salary is 186.7K. Just as in all professions salaries vary based on, region, company and years of experience.
